```python
# STEP 1: Load the Cleaned Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/NODVEDA DATASETS/cleaned_iris.csv")

df.head()
```

```python
# STEP 2: Understand Dataset Structure
df.info()
```

```python
df.describe()
```

```python
# STEP 3: Univariate Analysis (Histograms)
df.hist(figsize=(10,8))
plt.show()
```

```python
# Conclusion
# The sepal features appear approximately symmetric with moderate spread.
# The petal features display clear clustering patterns corresponding to different species.
# Minimal skewness and absence of extreme values suggest effective preprocessing.
# Petal measurements appear more discriminative than sepal measurements.
```

```python
# STEP 5: Scatter Plot (Relationship Between Features)
sns.scatterplot(x=df['petal_length'], y=df['petal_width'], hue=df['species'])
plt.show()
```

```python
# STEP 6: Correlation Matrix
corr = df.corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
#   Final EDA Summary You Should Be Able to Say

# The Iris dataset shows:

# Standardized numerical features centered around zero.

# Petal length and petal width are strongly correlated.

# Species classes are clearly separable using petal features.

# Sepal measurements provide weaker separation.

# No major outliers remain after preprocessing.

# This suggests petal measurements are the strongest predictors for classification.
```
